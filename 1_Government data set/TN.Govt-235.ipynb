```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))
df_cleaned['District Name'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Districts by Number of Water Bodies")
plt.xlabel("District Name")
plt.ylabel("Number of Water Bodies")
plt.xticks(rotation=45)
plt.show()
```

- Number of water bodies per district:The analysis visualizes the top 10 districts with the highest number of water bodies , highlighting Ramanathapuram and Sivagangai as the leading districts.

```python

plt.figure(figsize=(8, 8))
df_cleaned['water_body_ownership_name'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%', cmap='coolwarm')
plt.title("Top 5 Water Body Ownership Types")
plt.ylabel("")
plt.show()
```

- Distribution of water body ownership types:The analysis visualizes the top 5 water body ownership types, showing that Panchayat owns the majority (75.5%) of water bodies compared to other categories like Co-operative and Individual ownership.

```python

df_cleaned['storage_capacity_water_body_original'] = pd.to_numeric(df_cleaned['storage_capacity_water_body_original'], errors='coerce')

Q1 = df_cleaned['storage_capacity_water_body_original'].quantile(0.25)
Q3 = df_cleaned['storage_capacity_water_body_original'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df_cleaned[(df_cleaned['storage_capacity_water_body_original'] >= lower_bound) & 
                         (df_cleaned['storage_capacity_water_body_original'] <= upper_bound)]

plt.figure(figsize=(12, 6))
sns.histplot(df_filtered['storage_capacity_water_body_original'].dropna(), bins=30, kde=True, color='red', edgecolor='black')
plt.title("Distribution of Storage Capacity of Water Bodies (Without Outliers)")
plt.xlabel("Storage Capacity")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

- The analysis visualizes the distribution of water body storage capacities using a histogram with outliers removed using the IQR method, revealing a heavily right-skewed distribution where most water bodies have relatively low storage capacities.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_cleaned['longitude_dec'], y=df_cleaned['latitude_dec'], alpha=0.5, color='green')
plt.title("Geographical Distribution of Water Bodies")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

# Overview 
The dataset provides insights into water bodies in Tamil Nadu, covering their distribution, storage capacity, encroachment status, and community impact.  

# Key Findings
- Water Body Distribution: Most are in rural areas, concentrated in specific districts/blocks.  
- Storage Capacity:Many water bodies show reduced storage due to siltation and encroachment.  
- Encroachment & Fill-up Frequency:Encroachment affects a significant number, and many do not fill to capacity annually.  
- Community Impact: Certain villages and towns heavily depend on these water sources.  
- Water User Associations (WUAs): Present in some areas, but many lack governance structures.
   
# From the visualizations you provided, here are the key conclusions:

* The village TAYILGUNAMPATTANAM has the highest number of people benefiting from water bodies, followed by SENGUNDRAM and BODINAYAKANUR.
The distribution is significantly skewed, with a few villages benefiting disproportionately more than others.
This indicates that water bodies are not uniformly benefiting populations, possibly due to regional disparities in water resource management.
Basins with Most Water Bodies:

* The Cauvery Basin dominates, accounting for 63.8% of water bodies.
Other basins like the Barak Basin and regions in North Ladakh also contribute, but on a much smaller scale.
This suggests that water body distribution is geographically concentrated, which may influence regional water resource availability.
Most Mentioned Water Bodies:

* KULAM and KUTTAI are the most mentioned water bodies, indicating their prominence in local water management and reporting.
Traditional water bodies such as OORANI and ERI are also commonly referenced, highlighting the cultural and functional significance of these resources.

* A striking 81.4% of water bodies do not have a Water User Association.
Only 4.9% of water bodies are managed through WUAs, while 13.7% have unknown statuses.
This lack of formal management organizations may result in inefficient water usage, poor maintenance, and governance challenges
